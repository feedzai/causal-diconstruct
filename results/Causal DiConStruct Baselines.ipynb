```python
def sizeof_fmt(num, suffix="B"):
    for unit in ["", "K", "M", "G", "T"]:
        if abs(num) < 1024.0:
            return f"{num:3.1f} {unit}{suffix}"
        num /= 1024.0
    return f"{num:.1f} Yi{suffix}"
```

```python
FRAUD_NN_SAVE_BASE_PATH = "causal_diconstruct_fraudNN_baselines"
FRAUD_NN_ABS_SAVE_BASE_PATH = os.path.abspath(FRAUD_NN_SAVE_BASE_PATH)
FRAUD_NN_MODEL_RESULTS_PATH = os.path.join(FRAUD_NN_SAVE_BASE_PATH, "baseline_results.csv")

FRAUD_LGBM_SAVE_BASE_PATH = "causal_concept_distil_baselines"
FRAUD_LGBM_ABS_SAVE_BASE_PATH = os.path.abspath(FRAUD_LGBM_SAVE_BASE_PATH)
FRAUD_LGBM_MODEL_RESULTS_PATH = os.path.join(FRAUD_LGBM_SAVE_BASE_PATH, "baseline_results.csv")

CUB_SAVE_BASE_PATH = "causal_diconstruct_CUB_baselines"
CUB_ABS_SAVE_BASE_PATH = os.path.abspath(CUB_SAVE_BASE_PATH)
CUB_MODEL_RESULTS_PATH = os.path.join(CUB_SAVE_BASE_PATH, "baseline_results.csv")
```

```python
model_results_cub_df = pd.read_csv(CUB_MODEL_RESULTS_PATH)

model_results_fraud_lgbm_df = pd.read_csv(FRAUD_LGBM_MODEL_RESULTS_PATH)

model_results_fraud_nn_df = pd.read_csv(FRAUD_NN_MODEL_RESULTS_PATH)
```

```python
model_results_cub_df["independence_tradeof"] = (
    model_results_cub_df["test_mean_golden_roc_auc"]
    - (model_results_cub_df["test_independence_roc_auc"])
)

model_results_fraud_lgbm_df["independence_tradeof"] = (
    model_results_fraud_lgbm_df["test_mean_golden_roc_auc"]
    - (model_results_fraud_lgbm_df["test_independence_roc_auc"])
)

model_results_fraud_nn_df["independence_tradeof"] = (
    model_results_fraud_nn_df["test_mean_golden_roc_auc"]
    - (model_results_fraud_nn_df["test_independence_roc_auc"])
)
```
